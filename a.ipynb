````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
data=pd.read_csv('datasets.csv')
data.head()
````

````python
data.shape
````

````python
data.info()
````

````python
# statistical analysis
data.describe()
````

```Data Cleaning```

````python
data.isnull().sum()
````

````python
data.dropna(inplace=True)
data.isnull().sum()
````

````python
data.duplicated().sum()
data.drop_duplicates(inplace=True)
data.duplicated().sum()
````

````python
data.dtypes

data['id']=data['id'].astype(object)
data['host_id']=data['host_id'].astype(object)
data.dtypes
````

````python
# identifying outliers
sns.boxplot(data=data,x='price')
````

```Univariant Analysis```

````python
# price distribution
# removing outliers
df=data[data['price']<1500]
sns.histplot(data=df,x='price')
plt.ylabel('frequency')
````

````python
df.columns
````

````python
sns.histplot(data=data,x='availability_365')
plt.title('Distribution of Availability in 365 Days')
plt.xlabel('Availability in 365 Days')
plt.ylabel('Frequency')
````

````python
df.groupby(by='neighbourhood_group')['price'].mean()
````

````python
# price per bed
df['price_per_bed'] = df['price'] / df['beds']
df.head()
````

````python
df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
````

Bivariate Analysis

````python
sns.barplot(data=data,x='neighbourhood_group',y='price',hue='room_type')
````

````python
# number of review and price
sns.scatterplot(y='price', x='number_of_reviews', data=df,hue='neighbourhood_group')
````

````python
sns.pairplot(data=df,vars=['price','minimum_nights','availability_365','number_of_reviews'],hue='room_type')
````

````python
# geographical data
sns.scatterplot(x='longitude', y='latitude', data=df, hue='neighbourhood_group', palette="Set2")
````

````python
sns.scatterplot(x='longitude', y='latitude', data=df, hue='room_type', palette="Set2")
````

````python
# heat map chart
corr=df[['latitude','longitude','price', 'minimum_nights','number_of_reviews','reviews_per_month','availability_365','beds']].corr()
corr
sns.heatmap(data=corr,annot=True)
````
